# Step 1: Surrogate data generation

```python
mu_1 = 5
mu_a_true = mu_1
mu_2 = -6
mu_b_true = mu_2
sigma_1 = 1
sigma_a_true = sigma_1
sigma_2 = 2
sigma_b_true = sigma_2

x_1 = np.random.normal(mu_1, sigma_1, int(n/2))
x_2 = np.random.normal(mu_2, sigma_2, int(n/2))

list_x = []
list_x.extend(x_1)
list_x.extend(x_2)
```

# Step 2: Expectation-maximization method for the generative model of the mixture of two Gaussians in one dimension

```python
pil_im = display(Image.open('P(x_i|b).png'))
def probability_x_i_given_b(sigma_square, x, mu):
    
    result = (
        1 / math.sqrt(2 * math.pi * sigma_square)
    ) * math.exp(
        -1*(
            (x-mu)**2 / 2*sigma_square
        )
    )
    
    return result
```

```python
pil_im = display(Image.open('mu_a.png'))
pil_im = display(Image.open('mu_b.png'))
def mu_a_or_b(list_a, list_x):
    numerator = 0
    denominator = 0
    for index, element in enumerate(list_a):     
        numerator = numerator + (element * list_x[index])
        denominator = denominator + element
    mu = numerator / denominator
    
    return mu
```

```python
pil_im = display(Image.open('P(a|x_i).png'))
pil_im = display(Image.open('P(b|x_i).png'))
def probability_a_or_b_given_x_i(sigma_square_a, sigma_square_b, x, mu_a, mu_b, probability_a, probability_b):
    p_x_i_giv_b = probability_x_i_given_b(sigma_square=sigma_square_b, x=x, mu=mu_b)
    p_x_i_giv_a = probability_x_i_given_b(sigma_square=sigma_square_a, x=x, mu=mu_a)
        
    result = (p_x_i_giv_b * probability_b) / ((p_x_i_giv_b * probability_b) + (p_x_i_giv_a * probability_a))
    
    return result
```

```python
pil_im = display(Image.open('sigma_a^2.png'))
pil_im = display(Image.open('sigma_b^2.png'))
def estimated_sigma_square(list_a, list_x, mu):
    numerator = 0
    denominator = 0
    for index, element_a in enumerate(list_a):
        numerator = numerator + element_a * ((list_x[index] - mu)**2)
        denominator = denominator + element_a
        
    estimated_sigma_square = numerator / denominator
    
    return estimated_sigma_square
```

# Step 3: Plot pdf shapes and show convergence of parmeters by monitoring the values at each step

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

def plot_true_and_estimated_distributions(list_of_tuples_with_parameters):
    x_values = np.linspace(-10, 10, 120)
    for mu, sig in list_of_tuples_with_parameters:
        plt.plot(x_values, gaussian(x_values, mu[0], sig[0]), c='b')
        plt.plot(x_values, gaussian(x_values, mu[1], sig[1]), c='y')

    list_y = []
    for x_i in range(0, len(list_x)):
        list_y.append(0)
    
    plt.scatter(list_x, list_y, s=200)
```

```python
def expectation_maximization(list_x):
    # Initializations
    sigma_square_a = 0.2
    sigma_square_b = 0.2
    mu_a = -2
    mu_b = 2

    # Posterior probabilities (Bayes)
    probability_a = 0.5
    probability_b = 0.5
    
    fig, ax = plt.subplots(figsize=(15, 4))
    subplot_number = 0

    # Loop
    for iteration in range(1, max_iters + 1):
        subplot_number = subplot_number + 1
        subplot(1, max_iters + 1, subplot_number)
        plot_true_and_estimated_distributions(
            [
                ([mu_a_true, mu_a], [sigma_a_true, math.sqrt(sigma_square_a)]), 
                ([mu_b_true, mu_b], [sigma_b_true, math.sqrt(sigma_square_b)])
            ]
        )
        
        list_a = []
        list_b = []
        for index, i in enumerate(range(0, n)):
            # Coloring
            x_i = list_x[index]
            b_i = probability_a_or_b_given_x_i(sigma_square_a, sigma_square_b, x_i, mu_a, mu_b, probability_a, probability_b)
            list_b.append(b_i)
            list_a.append(1 - b_i)

        # Reestimation of Gaussian parameters
        mu_b = mu_a_or_b(list_b, list_x)
        sigma_square_b = estimated_sigma_square(list_a=list_b, list_x=list_x, mu=mu_b)

        mu_a = mu_a_or_b(list_a, list_x)
        sigma_square_a = estimated_sigma_square(list_a=list_a, list_x=list_x, mu=mu_a)
    
    subplot_number = subplot_number + 1
    subplot(1, max_iters + 1, subplot_number)
    plot_true_and_estimated_distributions(
        [
            ([mu_a_true, mu_a], [sigma_a_true, math.sqrt(sigma_square_a)]), 
            ([mu_b_true, mu_b], [sigma_b_true, math.sqrt(sigma_square_b)])
        ]
    )

    return mu_a, math.sqrt(sigma_square_a), mu_b, math.sqrt(sigma_square_b)

expectation_maximization(list_x)
```
