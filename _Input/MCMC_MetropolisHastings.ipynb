```python
#MCMC, Metropolis w/o Hastings
import numpy as np
import scipy.stats as st 
import seaborn as sns # Data Viz library 
import warnings # get rid of warnings:
warnings.filterwarnings("ignore") # Used for sns jointplot
# force to render all figures in the notebook
%matplotlib inline 

mus = np.array([5, 5])
sigmas = np.array([[1, .9], [.9, 1]])


def circle(x, y):
    return 0.1*(x-1)**2 + (y-2)**2 - 3**2


def pgauss(x, y):
    return st.multivariate_normal.pdf([x, y], mean=mus, cov=sigmas)


def metropolis_hastings(p, iter=100):
    x, y = 0., 0.
    samples = np.zeros((iter, 2))

    for i in range(iter):
        x_star, y_star = np.array([x, y]) + np.random.normal(size=2)
        if np.random.rand() < p(x_star, y_star) / p(x, y):
            x, y = x_star, y_star
        samples[i] = np.array([x, y])

    return samples

#Circle sampling
samples = metropolis_hastings(circle, iter=10000)
sns.jointplot(samples[:, 0], samples[:, 1]) #x=samples[:, 0]; y=samples[:, 1]

#Gaussian sampling
samples = metropolis_hastings(pgauss, iter=10000)
sns.jointplot(samples[:, 0], samples[:, 1])
```
